## Prediction for all routes using Prophet

```python
routes = train[['srcid','destid']].drop_duplicates()
```

```python
len(routes)
```

#### Adding Holidays

```python
newyr_lastweek = pd.DataFrame({
  'holiday': 'newyrweek',
  'ds': pd.to_datetime(['2023-12-31', '2024-12-31']),
  'lower_window': -7,
  'upper_window': 3,
})
repday = pd.DataFrame({
  'holiday': 'republicday',
  'ds': pd.to_datetime(['2023-01-26', '2024-01-26', '2025-01-26']),
  'lower_window': -1,
  'upper_window': 1,
})
indepday = pd.DataFrame({
  'holiday': 'independenceday',
  'ds': pd.to_datetime(['2023-08-15', '2024-08-15', '2025-08-15']),
  'lower_window': -1,
  'upper_window': 1,
})
diwali = pd.DataFrame({
  'holiday': 'diwali',
  'ds': pd.to_datetime(['2023-11-12', '2024-10-31', '2025-10-20']),
  'lower_window': -2,
  'upper_window': 1,
})
ugadi = pd.DataFrame({
  'holiday': 'ugadi',
  'ds': pd.to_datetime(['2023-03-22', '2024-04-09', '2025-03-30']),
  'lower_window': -1,
  'upper_window': 1,
})
onam = pd.DataFrame({
  'holiday': 'onam',
  'ds': pd.to_datetime(['2023-08-29', '2024-09-06', '2025-09-05']),
  'lower_window': -1,
  'upper_window': 1,
})
holidays = pd.concat((newyr_lastweek,repday,indepday,diwali,ugadi,onam))
```

```python
holidays
```

```python
for i in range(len(routes)):
    route = routes.iloc[i]
    print("Src : " ,route.srcid, ", Dest : ",route.destid)
    m = Prophet(holidays = holidays)
    
    m.add_seasonality(name='yearly', period=365.25, fourier_order=10)
    df = train[(train.srcid==route.srcid) & (train.destid==route.destid)]
    df = df.rename(columns = {'doj':'ds','final_seatcount':'y'})
    df['ds'] = pd.to_datetime(df['ds'])
    
    #Validation
    df_train1 = df[:-60]
    m.fit(df_train1)
    future1 = m.make_future_dataframe(periods=60)
    forecast1 = m.predict(future1)[['ds','yhat']]
    #forecast1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
    validation = forecast1[len(forecast1)-60:].merge(df[len(df)-60:],on='ds')
    print("RMSE : ",rmse(validation['y'],validation['yhat']))

    #Prediction
    m_pred = Prophet(holidays=holidays)
    m_pred.add_seasonality(name='yearly', period=365.25, fourier_order=10)
    m_pred.fit(df)
    future = m_pred.make_future_dataframe(periods=60)
    forecast = m_pred.predict(future)[['ds','yhat']]
    #forecast = forecast.rename(columns={'yhat':'final_seatcount'})
    #forecast1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
    suffix_src = str(route.srcid) if route.srcid>=10 else '0'+str(route.srcid)
    suffix_dest = str(route.destid) if route.destid>=10 else '0'+str(route.destid)
    forecast['route_key'] = forecast['ds'].astype(str) + '_' + suffix_src + '_' + suffix_dest
    #print(forecast['route_key'].head())
    sample1 = pd.merge(sample,forecast,on='route_key', how = 'left')
    sample['final_seatcount'] = sample1['yhat'].fillna(sample['final_seatcount'])
```
